# Treino de Analise de Dados
 - **Problematica:** Qual categoria de produtos tem o preço médio mais alto de compra?

```python
import pandas as pd
tabela = pd.read_csv("/content/dados_ecommerce.csv")

# Remoção de algumas colunas desnecessárias para análise da problemática
tabela = tabela.drop('ID', axis=1)
tabela = tabela.drop('Código de Barras', axis=1)
tabela = tabela.drop('URL da Imagem', axis=1)
tabela = tabela.drop('Local de Envio', axis=1)
tabela = tabela.drop('Promoção', axis=1)

tabela.head()
```

```python
print(tabela.describe())
```

```python
# Calcular a media por categoria
preco_medio_por_categoria = tabela.groupby('Categoria')['Preço'].mean().reset_index()

# Coloca a média presente no DataFrame em ordem decrescente
preco_medio_por_categoria = preco_medio_por_categoria.sort_values(by = 'Preço', ascending = False)
```

```python
# Visualização Gráfica
import matplotlib.pyplot as plt

plt.figure(figsize=(100,9))
plt.bar(preco_medio_por_categoria['Categoria'], preco_medio_por_categoria['Preço'])
plt.xlabel('Categoria')
plt.ylabel('Preço Médio de Compra')
plt.title('Preço Médio de Compra por Categoria')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()


plt.show()
```

```python
# Visualização Graáfica utilizando o plotly.express
import plotly.express as px

# Criação do Gráfico
graf = px.bar(preco_medio_por_categoria, x='Categoria', y='Preço', title='Preço Medio por Categoria',
              labels={'Categoria': 'Categoria', 'Preço': 'Preço Medio de Compra'})

# Rotacionar o nomes abaixo em 45° de cada linha para melhorar a visualização
graf.update_xaxes(tickangle=45, tickmode='array', tickvals=list(preco_medio_por_categoria))

graf.show()
```

# Conclusões
- A partir de uma analise basica utilizando o Python juntos com suas biblioteca de analise e visualização gráfica tornou-se fácil a percepção do preço medio mais alto das categorias.
- Olhando o grafico acima da para perceber que o preço medio mais alto de produtos está com a categoria "quam".
